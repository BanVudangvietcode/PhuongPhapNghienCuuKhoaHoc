 Histogram – Biểu đồ tần suất (phân phối rời rạc)

```python
from matplotlib import pyplot
import pandas as pd

filename = 'diabetes.csv'
data = pd.read_csv(filename)  # KHÔNG cần names=

print(data.dtypes)  # Kiểm tra kiểu dữ liệu

data.hist()
pyplot.show()
```

2. Density Plot – Biểu đồ mật độ (phân phối mượt)

```python
import pandas as pd
from matplotlib import pyplot

filename = 'diabetes.csv'
data = pd.read_csv(filename)

# Vẽ biểu đồ mật độ
data.plot(kind='density', subplots=True, layout=(3, 3), sharex=False)
pyplot.show()
```

3. Box and Whisker Plot – Biểu đồ hộp

```python
import pandas as pd
from matplotlib import pyplot

filename = 'diabetes.csv'
data = pd.read_csv(filename)

# Vẽ biểu đồ hộp
data.plot(kind='box', subplots=True, layout=(3, 3), sharex=False, sharey=False)
pyplot.show()
```

4. Correlation Matrix Plot – Ma trận tương quan có nhãn

```python
import pandas as pd
from matplotlib import pyplot
import numpy as np

filename = 'diabetes.csv'
data = pd.read_csv(filename)

# Tính toán ma trận tương quan
correlations = data.corr()

# Vẽ ma trận tương quan
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

# Gán nhãn trục
ticks = np.arange(0, len(data.columns), 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(data.columns, rotation=90)
ax.set_yticklabels(data.columns)

pyplot.show()
```

5. Correlation Matrix Plot – Không nhãn

```python
import pandas as pd
from matplotlib import pyplot

filename = 'diabetes.csv'
data = pd.read_csv(filename)

correlations = data.corr()

fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

pyplot.show()
```
